**Sebelum melakukan pendeklarasian dengan grafik kita harus mengimport package dibawah ini**

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
bankMarketing = pd.read_csv('bank-additional-full.csv',sep=";") #pendeklarasian untuk membaca data
bankMarketing.head()
```

**1. Grafik Batang**

Grafik batang ini menunjukkan umur pelanggan dan status pekerjaan pelanggan. 


```python
x = list (bankMarketing["age"])
y = list (bankMarketing["job"])

plt.figure(figsize =(7, 10))
plt.bar(x, y, color='blue')
plt.xlabel("AGE")
plt.ylabel("JOB")
plt.title("umur dan pekerjaan pelanggan")
plt.show()
```

**2. Pie Chart**


Grafik pie dibawah ini menunjukkan bahwa pelanggan yang melakukan peminjaman di bank ada sebanyak 15.2%, yang tidak melakukan peminjaman ada sebanyak 82.4% dan sisanya yaitu 2.4% merupakan nilai null pada data tersebut.

```python
data_pie = bankMarketing['loan'].value_counts().rename('loan').reset_index(name='id_count') 
data_pie
```

***

```python
plt.pie(data_pie.id_count, labels=data_pie["index"], startangle=90,
autopct='%.1f%%', shadow= True, explode= (0.2, 0.1 , 0))
plt.title('pinjaman')
plt.show()
```

**3. Histogram**


Dari analisis distribusi usia nasabah, kita dapat menyimpulkan bahwa bank memiliki mayoritas nasabah muda, sementara jumlah nasabah yang lebih tua terbatas.

```python
plt.figure(figsize=(10, 6))
plt.hist(bankMarketing['age'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribusi Usia Nasabah')
plt.xlabel('Usia')
plt.ylabel('Jumlah Nasabah')
plt.show()
```

**4. Heatmap**

Dari heatmap korelasi antara variabel numerik dalam dataset bank, dapat diambil kesimpulan bahwa terdapat beberapa hubungan yang signifikan antara variabel-variabel tersebut.


```python
correlation_matrix = bankMarketing.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Heatmap Korelasi Antara Variabel Numerik')
plt.xlabel('Variabel')
plt.ylabel('Variabel')
plt.show()
```

**5. Box Plot**

Grafik diatas menunjukkan data perumahan dan data periklanan, ada beberapa yang memiliki dan ada beberapa yang tidak memiliki perumahan dan periklanan untuk mengetahui bank ini, dan ada yang tdak tau

```python
sns.boxplot(data = bankMarketing, x = 'housing', y = 'campaign')
```
