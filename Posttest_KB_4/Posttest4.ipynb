**Import Library**



```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, StandardScaler, OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
```

**Info Data**

**1. Membagi dataset menjadi training set dan testing set dengan proporsi 70:30**

```python
X = df.iloc[:, : -1]
y = df.iloc[:, -1]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

print("Dimensi X_train", X_train.shape)
print("Dimensi X_test", X_test.shape)
print("Dimensi y_train", y_train.shape)
print("Dimensi y_test", y_test.shape)
```

**2. Melakukan normalisasi data pada salah satu attribute menggunakan Min Max scaler dengan mengcopy dataset terlebih dahulu**

```python
min_max = MinMaxScaler()

# Copy Data
copy1 = df
dataCopy_normalization = copy1[["duration"]]

normalization = min_max.fit_transform(dataCopy_normalization[["duration"]])

data_normalization = pd.DataFrame(normalization)
data_normalization.columns =["duration_normalisasi"]

dataCopy_normalization = dataCopy_normalization.join(data_normalization["duration_normalisasi"]).reset_index()
dataCopy_normalization = dataCopy_normalization.drop("index", axis=1)

print()
print("Data Sebelum dan Sesudah dilakukan normalisasi : \n")
dataCopy_normalization.head(10)
```

**3. Melakukan standarisasi pada dataset dengan membuat copy dataset terlebih dahulu**

**4. Lakukan Data cleaning pada data dengan nilai null dengan bilangan bulat : median/modus, bilangan desimal : mean, tulisan : modus**

- Mengecek nilai null

```python
df.isnull().sum()
```

- Karena Data tidak ada yang bernilai null maka dibuat

```python
# Untuk Menghilangkan SettingWithCopyWarning
pd.options.mode.chained_assignment = None

df.age[[15,100,200,300,400,500,600,700,800,900,1000]]=None;
df.euribor3m[[15,100,200,300,400,500,600,700,800,900,1000]]=None;
df.job[[15,100,200,300,400,500,600,700,800,900,1000]]=None;

df[df.isnull().any(axis=1)]
```

- Data Bilangan Bulat dengan Strategi Median

```python
strategi1 = SimpleImputer(strategy="median")

df["age"] = strategi1.fit_transform(df[["age"]])

df.isna().sum()
```

```python
print(df.age[[15,100,200,300,400,500,600,700,800,900,1000]])
```

- Data Bilangan Desimal dengan Strategi Mean

```python
strategi2 = SimpleImputer(strategy="mean")

df["euribor3m"] = strategi2.fit_transform(df[["euribor3m"]])

df.isna().sum()
```

```python
print(df.euribor3m[[15,100,200,300,400,500,600,700,800,900,1000]])
```

- Data Tulisan dengan Strategi Modus

```python
strategi3 = SimpleImputer(missing_values=None, strategy="most_frequent")

df["job"] = strategi3.fit_transform(df[["job"]])

df.isna().sum()
```

```python
print(df.job[[15,100,200,300,400,500,600,700,800,900,1000]])
```
