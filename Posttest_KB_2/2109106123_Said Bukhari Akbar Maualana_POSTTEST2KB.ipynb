**3. Tampilkan informasi jumlah record, jumlah attribute, nama attribute, jumlah attribute dengan tipe angka. Masukkan informasi tersebut dalam dictionary dengan key sesuai nama informasinya. Tampilkan isi dari dictionary tersebut sebagus mungkin.**

```python
# Mendapatkan jumlah record
jumlah_record = len(bankMarketing)

# Mendapatkan jumlah attribute
jumlah_attribute = len(bankMarketing.columns)

# Mendapatkan nama-nama attribute
nama_attribute = bankMarketing.columns.tolist()

# Mendapatkan jumlah attribute dengan tipe angka
jumlah_attribute_angka = len(bankMarketing.select_dtypes(include=['number']).columns)

# Membuat dictionary dengan informasi yang diperlukan
informasi = {
    'Jumlah Record': jumlah_record,
    'Jumlah Attribute': jumlah_attribute,
    'Nama Attribute': nama_attribute,
    'Jumlah Attribute Angka': jumlah_attribute_angka
}

informasi
```

***4.  Tampilkan nilai dataset pada baris dimulai dengan 3 digit NIM akhir hingga data terakhir dan kolom ke 4 hingga kolom terakhir menggunakan ke 2 metode slice.***

- Menggunakan iloc untuk slicing

```python
# Menggunakan iloc untuk slicing
bankMarketing = pd.read_csv('bank-additional-full.csv',sep=";")
metode1 = bankMarketing.iloc[123:, 3:]
metode1
```

- Menggunakan loc untuk slicing

```python
# Menggunakan loc untuk slicing
metode2 = bankMarketing[123:].iloc[:, 3:]
metode2
```
